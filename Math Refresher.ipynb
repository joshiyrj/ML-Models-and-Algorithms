```python
import numpy as np
from numpy import linalg as LA
x = np.random.random()
Q = np.array([[np.cos(x), -np.sin(x)],[np.sin(x),np.cos(x)]])
LA.norm(Q[0,:]),LA.norm(Q[1,:]),np.dot(Q[0,:],Q[1,:])
```

```python
D = np.diag((-1,1))
LA.eigvals(D)
```

```python
A = np.dot(Q,D)
A = np.dot(A,Q.T)
LA.eigvals(A)
```
