```python
d = {'one' : pd.Series([1,2,3], index=['a','b','c']),
    'two' : pd.Series([4,5,6,7],index=['a','b','c','d']),
    'three' : pd.Series([5,20,3],index = ['a','b','c'])  }


x = pd.DataFrame(d)
print("Our dataframe is: ")
print(x)

print("using Del function")
del x['one']
print(x)

print("Deleting column using POP function: ")
x.pop('two')
print(x)
```
