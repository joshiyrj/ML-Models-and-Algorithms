```python
def binary_step(x):
  if x>0:
    return 0
  else:
    return 1
binary_step(5), binary_step(-1)
```

```python
def linear_function(x):
  return 4*x
linear_function(4), linear_function(-2)
```

```python
import numpy as np
def sigmoid_function(x):
  z = (1/(1 + np.exp(-x)))
  return z
sigmoid_function(7), sigmoid_function(-22)
```

```python
def tanh_function(x):
  z = (2/(1 + np.exp(-2*x))) -1
  return z
tanh_function(0.5), tanh_function(-1)
```

```python
def relu_function(x):
  if x<0:
    return 0
  else:
    return x
relu_function(7), relu_function(-7)
```

```python
def leaky_relu_function(x):
  if x<0:
    return 0.01*x
  else:
    return x
leaky_relu_function(7), leaky_relu_function(-7)
```

```python
def elu_function(x, a):
  if x<0:
    return a*(np.exp(x)-1)
  else:
    return x
elu_function(5, 0.1),elu_function(-5,0.1)
```

```python
def swish_function(x):
  return x/(1-np.exp(-x))
swish_function(-67), swish_function(4)
```

```python
def softmax_function(x):
  z = np.exp(x)
  z_ = z/z.sum()
  return z_
softmax_function([0.8,1.2,3.1])
```
